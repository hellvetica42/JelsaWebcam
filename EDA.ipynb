```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('logged_data.csv')
df
```

```python
df[df.Class == 'person']
```

```python
def plot_camera_class(cam_name, class_name):
    filtered_df = df[(df.Name == cam_name) & (df.Class == class_name)]
    filtered_df.loc[:, 'timestamp'] = pd.to_datetime(filtered_df.Timestamp)
    filtered_df.loc[:, 'time'] = filtered_df['timestamp'].dt.time
    plt.figure(figsize=(12, 8))
    plt.plot_date(filtered_df['timestamp'], filtered_df.Detections)
    plt.gcf().autofmt_xdate()
    plt.xlabel('Time (HH:MM)')
    plt.ylabel('Value')
    plt.title('Value over time')
    plt.tight_layout()
    plt.show()
```

```python
plot_camera_class('Pjaca', 'person')
```
